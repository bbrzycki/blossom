```python
organism1 = [[0], 0.0]
```

```python
organism1
```

```python
import numpy as np
import random
```

```python
total_time = 50000000
timestep = 1.0
positions = []
# print("Initial:", organism1)
for i in np.arange(0, total_time, timestep):
    # change position randomly
    if random.random() < 0.5:
        organism1[0][0] -= 1
    else:
        organism1[0][0] += 1
        
    # increment organism's age
    organism1[1] += timestep
    
    # add position to the list
    positions.append(organism1[0][0])
    
    # display organism information
    # print("t=%s:" % organism1[1], organism1)
    
```

```python
# positions
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(40, 10))
plt.plot(positions)
plt.ylabel('position')

plt.show()
```

```python
from scipy.fftpack import fft
```

```python
# Number of samplepoints
N = len(positions)
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)

yf = fft(positions)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

plt.figure(figsize=(40, 10))
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))

axes = plt.gca()
axes.set_xlim([0,10])
axes.set_ylim([0,0.1])

plt.show()
```
